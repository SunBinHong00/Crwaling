```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

word = '명언'
page = 1

eng_list = []
kr_list = []
```

```python
wise_list = ['인생명언','인내명언','독서명언','친구명언','정직명언','실패명언','돈명언','컴퓨터명언','언어명언','명예명언',
'예술명언','성공명언','생각명언','공부명언','지혜명언','시간명언','일명언','용기명언','희망명언','도전명언','건강명언','원수명언',
'용서명언','음식명언','자유명언','경험명언','계절명언','겸손명언','믿음명언','자기관리명언','정의명언','침묵명언','사랑명언','노력명언',
'행복명언','자신감명언','아름다움명언','죽음명언','음악명언','가족명언','이기심명언','남녀명언','전쟁명언','고독명언','습관명언','스포츠명언',
'지식명언','과학명언','동물명언','행운명언','이별명언','후회명언']
```

```python
browser = webdriver.Chrome()
for word in wise_list:
    for page in range(1,135):
        url = f'https://en.dict.naver.com/#/search?range=example&page={page}&query={word}&shouldSearchVlive=false&autoConvert='
        browser.get(url)

        # Wait for the English elements to be present
        try:
            WebDriverWait(browser, 3).until(
                EC.presence_of_all_elements_located((By.XPATH, '//span[@class="text"]'))
            )
        except:
            break

        # Wait for the Korean elements to be present
        WebDriverWait(browser, 3).until(
            EC.presence_of_all_elements_located((By.XPATH, '//p[@class="text"]'))
        )

        eng = browser.find_elements(By.XPATH, '//span[@class="text"]')[::2]
        kr = browser.find_elements(By.XPATH, '//p[@class="text"]')

        eng_list.extend([e.text for e in eng])
        kr_list.extend([k.text for k in kr])
        
len(kr_list), len(eng_list)
```

```python
import pandas as pd

df = pd.DataFrame({'kr':kr_list,
              'eng':eng_list})

# ~~명언의 형식을 띄지않는 데이터 drop
drop_idx = df['kr'].str.extract(r'(\w+명언)')[0].isna()
wise_df = df[~drop_idx].reset_index(drop=True)
wise_list = wise_df['kr'].str.extract(r'(\w+명언)')[0].unique()
[print("'"+i+"'") for i in wise_list]
```
